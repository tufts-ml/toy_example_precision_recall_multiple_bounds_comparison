```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
proba_grid = np.linspace(1e-5, 1, 1000)
```

```python
proba_grid
```

```python
ce_for_pos_example = np.log2(proba_grid)
error_rate = 1.0*(proba_grid<=0.5)

f, axs = plt.subplots(1, 1)
axs.plot(proba_grid, -ce_for_pos_example, label='log loss')
axs.plot(proba_grid, error_rate, '--', label='error rate (1-accuracy)')
axs.set_xlabel('Probability')
# axs.set_ylabel('Cross entropy')
axs.set_title('true label = 1')
axs.legend()
plt.show()
```
